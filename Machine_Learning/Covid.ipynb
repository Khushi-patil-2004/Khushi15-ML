```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_excel("covid_dataset_500_entries.xlsx")
data.head()
```

```python
data.shape
```

```python
data.info
```

```python
data.isnull().sum()
```

```python
data['Actual']=data['Actual'].map({'Positive':1,'Negative':0})
data['Predicted']=data['Predicted'].map({'Positive':1,'Negative':0})
```

```python
data.head()
```

```python
X =data['Actual']
Y = data['Predicted']
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy=accuracy_score(X,Y)
precision=precision_score(X,Y)
recall=recall_score(X,Y)
f1=f1_score(X,Y)
```

```python
print("Accuracy score is:",accuracy)
print("Precision score is:",precision)
print("Recall score is:",recall)
print("F1 score is:",f1)
```

```python
tn, fp, fn, tp = confusion_matrix(X,Y).ravel()
```

```python
conf_matrix = np.array([[tn, fp], [fn, tp]])
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix,annot=True,cmap='Blues',fmt='d', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.show()
```

```python

print("\nConfusion Matrix Results:")
print(f"True Positives: {tp}")
print(f"False Positives: {fp}")
print(f"True Negatives: {tn}")
print(f"False Negatives: {fn}")
```
